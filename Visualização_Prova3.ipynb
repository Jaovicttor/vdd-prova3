```python
import pandas as pd
```

```python
df = pd.read_csv('/content/brasil-mineracao.csv')
```

```python
df.head()
```

```python
df_filtrado = df.drop(['PROCESSO', 'NUMERO', 'ID', 'ULT_EVENTO', 'DSProcesso', 'FASE'], axis=1)
```

```python
df_filtrado = df_filtrado[df_filtrado['SUBS'] != 'DADO NÃO CADASTRADO']
```

```python
df_filtrado.head()
```

```python
categorias = {
    "Metais-Ferrosos": ["FERRO", "MINÉRIO DE FERRO", "HEMATITA", "MAGNETITA", "ITABIRITO"],
    "Metais-Não-ferrosos": ["ALUMÍNIO", "MINÉRIO DE ALUMÍNIO", "BAUXITA", "COBRE", "NÍQUEL", "ZINCO", "CHUMBO"],
    "Metais Preciosos": ["OURO", "PRATA", "PALÁDIO", "PLATINA", "MINÉRIO DE OURO", "OURO NATIVO"],
    "Metais Pesados": ["CROMO", "CÁDMIO", "BISMUTO", "VANÁDIO", "MOLIBDÊNIO", "ESTANHO", "ANTIMÔNIO"],
    "Terras Rara": ["NIÓBIO", "TÂNTALO", "TERRAS RARAS", "MONAZITA", "RUTILO", "ZIRCÔNIO", "HÁFNIO", "MINÉRIO DE CÉRIO"],
    "Rochas de Construção/Ornamentais": ["GRANITO", "MÁRMORE", "GNAISSE", "QUARTZITO", "BASALTO", "ARDÓSIA", "MIGMATITO"],
    "Rochas de Agregados para Brita": ["BRITA DE GRANITO", "SEIXOS", "PEDREGULHO", "CASCALHO", "AREIA", "SAIBRO"],
    "Minerais Industriais": ["CALCÁRIO", "DOLOMITO", "GIPSITA", "ARGILA", "CAULIM", "FELDSPATO", "TALCO", "SÍLICA", "MICA"],
    "Químicos / Fertilizantes": ["FOSFATO", "SALGEMA", "ENXOFRE", "SAIS DE POTÁSSIO", "SILVINITA", "ROCHA POTÁSSICA", "CARNALITA"],
    "Gemas / Minerais Preciosos": ["DIAMANTE", "AMETISTA", "TOPÁZIO", "TURMALINA", "ESMERALDA", "SAFIRA", "RUBI", "OPALA", "CITRINO"],
    "Hídricos": ["ÁGUA MINERAL", "ÁGUAS TERMAIS", "ÁGUA POTÁVEL DE MESA", "ÁGUA TERMO MINERAL"],
    "Energéticos": ["CARVÃO MINERAL", "ANTRACITO", "TURFA", "LINHITO", "XISTO BETUMINOSO", "ROCHA BETUMINOSA"]
}
```

```python
def mapear_categoria(subs):
    for categoria, substancias in categorias.items():
        if subs.upper() in substancias:
            return categoria
    return None
```

```python
df_filtrado['CATEGORIA'] = df_filtrado['SUBS'].apply(mapear_categoria)
```

```python
df.head()
```

```python
df_filtrado = df_filtrado[df_filtrado['CATEGORIA'].notna()]
```

```python
df_filtrado
```

```python
df_ordenado = df_filtrado.sort_values(by='ANO', ascending=False)
df_ordenado
```

```python
colunas = ['ANO'] + [col for col in df_ordenado.columns if col != 'ANO']
df_ordenado = df_ordenado[colunas]
```
