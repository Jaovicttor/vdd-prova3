```python
categorias = {
    "Gemas/Pedras Preciosas": [
        'DIAMANTE', 'ESMERALDA', 'RUBI', 'SAFIRA', 
        'TURMALINA', 'AMETISTA', 'TOPÁZIO', 'ALEXANDRITA', 'ÁGATA',
        'OPALA', 'TURQUESA'
    ],
    
    "Metais Pesados": ['CHUMBO', 'CROMO', 'CÁDMIO', 'ZINCO', 'MERCÚRIO'
    ],

    "Metais Preciosos": ['OURO', 'PRATA', 'PLATINA'
    ],
    
    "Metais Estratégicos": [
        'TUNGSTÊNIO', 'NÍQUEL', 'ALUMÍNIO',
        'FERRO', 'COBRE', 'MANGANÊS', 'TITÂNIO', 'ESTANHO', 'LÍTIO'
    ],
    
    "Minerais Industriais": [
        'QUARTZO', 'FELDSPATO', 'TALCO', 'GRAFITA', 'CAULIM', 'MICA',
        'ENXOFRE', 'GIPSITA', 'BARITA', 'FLUORITA'
    ],
    
    "Rochas de Construção": [
        'GRANITO', 'MÁRMORE', 'BASALTO', 'ARDÓSIA', 'ARENITO',
        'GNAISSE', 'CALCÁRIO'
    ],
    
    "Recursos Energéticos": [
        'CARVÃO', 'URÂNIO'
    ],
    
    "Elementos/Compostos Químicos": [
        'FOSFATO', 'SÍLICA', 'BORATOS', 'MAGNÉSIO'
    ],
    
    "Agregados/Sedimentos": [
        'AREIA', 'ARGILA', 'CASCALHO', 'SAIBRO'
    ],

    "Recursos Hídricos": [
        'ÁGUA MINERAL'
    ],
}
```

```python
def mapear_categoria(subs):
    for categoria, substancias in categorias.items():
        if subs.upper() in substancias:
            return categoria
    return None
```

```python
df_filtrado['CATEGORIA'] = df_filtrado['SUBS'].apply(mapear_categoria)
```
